### <center>1. Gassmann Equations</center>

The Gassmann equations were derived by Fritz Gassmann in 1951. Gassmann's equations are a set of two equations describing the isotropic elastic constants of an ensemble consisting of an isotropic, homogeneous porous medium with a fully connected pore space, saturated by a compressible fluid at pressure equilibrium. Clasically in the hydrocarbon industry, the Gassmann equations are commonly used to model the change to the elastic properties of a rock undergoing fluid substitution, that is, <b>calculating the change in the elastic properties of a bulk rock when a rock is saturated with another fluid</b>. The Gassmann equations is defined by (Mavko et al., 2009):

$K_{sat} = K_{dry} + \frac{\left(1-\frac{K_{dry}}{K_{matrix}}\right)^{2}}{\frac{\phi}{K_{fluid}}+\frac{(1-\phi)}{K_{matrix}}-\frac{K_{dry}}{K^{2}_{matrix}}}, \quad \mu_{sat}=\mu_{dry}$

Where $K_{sat}$ is the effective bulk modulus of the saturated rock, $K_{dry}$ is the effective bulk modulus of the dry rock frame, $K_{matrix}$ is the rock matrix bulk modulus, $K_{fluid}$ is the fluid bulk modulus, $\phi$ is the porosity, and $\mu_{sat}$ and $\mu_{dry}$ are the shear moduli of the saturated rock and dry rock frame respectively.

The Gassmann equations have the following assumptions/limitations:
- The mineral modulus is homogeneous and isotropic
- All pore spaces within the rock are interconnected and of the same geometry
- The pore fluid is frictionless and inert and does not cause any chemical reactions within the rock frame
- Wave-induced pressure changes throughout the pore space have equilibrated

In practice, these assumptions don’t reflect what occurs in nature and can limit the nature of experimental work
(e.g. Han and Batzle, 2004). As Gassmann’s assumptions allow pressure equilibration, this implies validity only at low frequencies.
However, these assumptions are assumed to hold for unconsolidated clastic rocks at low seismic frequencies (<100Hz)
(Daley, 2009), where situationally the pore-fluid has zero rigidity. At higher frequencies, Wave-Induced Fluid Flow (WIFF) mechanisms persist, where
a propagating seismic wave induces differential pore pressure gradients, causing complex fluid flow. This relative
motion of fluid between pores leads to velocity dispersion and attenuation . WIFF’s have no spatial scale-
dependence, and can occur from micro- to mesoscale and a temporally-dependent on the relaxation period of the
induced pressure gradients. The interconnected and geometrically similar pore shape disregards the presence of isolated pores, microcracks, cracks and fractures too which are pronounced as anisotropy in the seismic domain.

### <center>2. Effective Medium Theory's</center>

Effective Medium Theory's (EMTs) are theoretical formulations used to model the macroscopic elastic properties of a composite rock on the basis of its microscopic constituents. In essence, they assume that a whole rock is made up of constituent materials (solids, fluids and gases) and that each of these components posess their own respective properties which contribute as a whole to the observable properties of the bulk rock. Below I will expand upon the non-Gassmann models listed in <b>Figures 2 and 3</b> from the PGR Conference poster.

#### <center>2.1. Hudson's Model for Cracked Media</center>

Hudson's model is based on a scattering-theory analysis of the mean wavefield in an elastic solid with thin, penny-shaped ellipsoidal cracks or inclusions (<i>Hudson, 1980; 1981</i>). The effective moduli $C^{eff}_{ijkl}$ is given by:
$$ C^{eff}_{ijkl} = C^{0}_{ij} + C^{1}_{ij} + C^{2}_{ij}$$
Where $C^{0}_{ij}$ are the isotropic background moduli, and $C^{1}_{ij}$ and $C^{2}_{ij}$ are the first- and second-order corrections, respectively. For a single crack set with cracks aligned normal to the 3-axis, the cracked media show transverse isotropic symmetry, and the corrections are:
$$C^{1}_{11} = -\frac{\lambda^{2}}{\mu} \varepsilon U_{3}$$
$$C^{1}_{13} = -\frac{\lambda(\lambda + 2\mu)}{\mu} \varepsilon U_{3}$$
$$C^{1}_{33} = -\frac{(\lambda + 2\mu)^{2}}{\mu} \varepsilon U_{3}$$
$$C^{1}_{44} = -\mu \varepsilon U_{1}$$
$$C^{1}_{66} = 0$$
$$C^{2}_{11} = \frac{q}{15}\frac{\lambda^{2}}{(\lambda + 2\mu)}(\varepsilon U_{3})^{2}$$
$$C^{2}_{13} = \frac{q}{15} \lambda (\varepsilon U_{3})^{2}$$
$$C^{2}_{33} = \frac{q}{15} (\lambda + 2\mu)(\varepsilon U_{3})^{2}$$
$$C^{2}_{44} = \frac{2}{15}\frac{\mu (3\lambda + 8\mu)}{\lambda + 2\mu}(\varepsilon U_{1})^{2}$$
$$C^{2}_{66} = 0$$
Where:
$$q = 15\frac{\lambda^{2}}{\mu^{2}}+28\frac{\lambda}{\mu}+28$$
$$\varepsilon = \frac{N}{V}a^{3} = \frac{3\phi}{4\pi\alpha} = \text{crack density}$$
The isotropic background moduli are $\lambda$ and $\mu$, and $a$ and $\alpha$ are the crack radius and aspect ratio, respectively. The corrections $C^{1}_{ij}$ and $C^{2}_{ij}$ obey the usual symmetry properties for transverse isotropy or hexagonal symmetry.<br><br>
> Note: The second-order expansion is not a uniformly converging series and predicting increasing moduli with crack density beyond the formal limit (<i>Cheng, 1993</i>). Better results will be obtained by using just the first-order correction rather than inappropriately using the second-order correction. Cheng gives a new expansion based on the Pade´ approximation, which avoids this problem. 

The terms $U_{1}$ and $U_{3}$ vary depending upon the crack conditions. For dry cracks:
$$U_{1} = \frac{16(\lambda + 2\mu)}{3(3\lambda + 4\mu)} \; ; \; U_{3} = \frac{4(\lambda + 2\mu)}{3(\lambda + \mu)}$$
For "weak" inclusions (i.e. When $\frac{\mu\alpha}{(K' + \frac{4}{3}\mu)}$ is out the order of 1 and is not small enough to be neglected):
$$U_{1} = \frac{16(\lambda + 2\mu)}{3(3\lambda + 4\mu)}\frac{1}{(1+M)} \; ; \; U_{3} = \frac{4(\lambda + 2\mu)}{3(\lambda + \mu)}\frac{1}{(1 + \kappa)}$$
Where:
$$M = \frac{4 \mu'}{\pi\alpha\mu}\frac{(\lambda + 2\mu)}{3\lambda + 4\mu} \; ; \; \kappa = \frac{(K' + \frac{4}{3}\mu')(\lambda + 2\mu)}{\pi\alpha\mu(\lambda + \mu)}$$
Where $K'$ and $\mu'$ are the bulk and shear modulus of the inclusion material. The criteria for an inclusion to be "weak" depend on its shape or aspect ratio $\alpha$ as well as on the relative moduli of the inclusion and the matrix material. Dry cavities can be modelled by setting the inclusion moduli to zero. Fluid-saturated cavities are simulated by setting the inclusion shear modulus to zero.<br><br>
> Because the cavities are isolated with respect to flow, this approach simulates very high-frequency behavior appropriate to ultrasonic laboratory conditions. At low frequencies, when there is time for wave-induced pore-pressure increments to flow and equilibrate, it is better to find the effective moduli for dry cavities and then saturate them with the Brown and Korringa low-frequency relations. This should not be confused with the tendency to term this approach a low-frequency theory, for crack dimensions are assumed to be much smaller than a wavelength.

#### <center>2.2. Standard Linear Solid Model</center>

The Standard Linear Solid (SLS) model, or the Zener model, is a viscoelastic representation of a dashpot for a material which is both viscous and elastic. In Rock Physics Modelling, this is a frequency-dependent, saturation-independent, isotropic model which can be used to model a viscoelastic/anelastic rock. This means that the after energy is passed to the rock, the formation creeps and relaxes as a function of the viscosity of the saturating fluid, causing energy loss which we know as <b>attenuation</b> (Ursin and Toverud, 2002).

#### <center>2.6. Frequency-Dependent Anisotropic Partial Saturation Model</center>

The Frequency-Dependent Anisotropic Partial Saturation Model (Jin et al., 2018) is an extension of the work of Papageorgiou and Chapman (2017) which builds upon the work of Chapman et al. (2002). The foundational work of Chapman et al. (2002) builds upon the Eshelby (1959) model of penny-shaped cracks and inclusions to account for the frequency-dependent, wave-induced exchange of fluids between pores and cracks, as well as between cracks of different orientations. The model both agrees with the Gassmann equations at low (<100 Hz) and squirt flow related dispersion at higher frequencies. Papageorgiou and Chapman (2017) builds upon the original model to include the partial saturation effects on the fluid modulus of two immiscible fluids as a function of the dimensionless capillary pressure parameter, <u>q</u>. This was further developed by Jin et al. (2018) to derive the frequency-dependent bulk and shear modulus, given by:

$K_{eff}(\omega) = \lambda+\frac{2}{3}\mu-\frac{4}{3}\varepsilon\frac{\left(\lambda+\frac{2}{3}\mu \right)^{2}(\lambda+2\mu)}{\mu(\lambda+\mu)}-\phi\frac{\left(\lambda+\frac{2}{3}\mu\right)(\lambda+2\mu)}{4\mu} \\
    +\varepsilon\left[\frac{4\left(\lambda+\frac{2}{3}\mu\right)^{2}(\lambda+2\mu)}{3\mu(\lambda+\mu)}+4\pi r\left(\lambda+\frac{2}{3}\mu\right)\right]A(\omega) \\
    +\phi\frac{9\left(\lambda+\frac{2}{3}\mu\right)(\lambda+2\mu)}{4\mu}B(\omega)$

$\mu_{eff}(\omega)=\mu-\frac{16}{45}\varepsilon\frac{1}{1+K_{c}}\frac{\mu(\lambda+2\mu)}{3\lambda+4\mu}\left(K_{c}+\frac{1}{1+i\omega\tau}\right) \\
    -\frac{32}{45}\varepsilon\frac{\mu(\lambda+2\mu)}{3\lambda+4\mu}-\phi\frac{15\mu(\lambda+2\mu)}{9\lambda+14\mu}$

Where $\omega$ is the angular frequency, $\mu$ and $\lambda$ are the shear modulus and Lam\'{e} parameter of the rock matrix, $\varepsilon$ is the crack density, $r$ is the crack aspect ratio, $\phi$ is the total porosity and $\tau$ is the timescale parameter, which controls the frequency regime over which dispersion occurs, which is given by:

$\tau \approx \frac{4\eta a^{3}(1-\nu)}{9\kappa\zeta\mu}$

```python
%load_ext autoreload
%autoreload 2
import numpy as np
import mantis_core.rock_physics as manRP
import mantis_core.rock_physics.fluid as manFL
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
plt.style.use("./mpsplot_default.mplstyle")
import numpy as np
```

```python
water = manFL.Fluid.from_presets(name="Water", temperature=40, pressure=14)
co2 = manFL.Fluid.from_presets(name="CarbonDioxide", temperature=40, pressure=14)
brine = manFL.Fluid(name="Brine", modulus = 3.59, viscosity = 0.000932720, density = 1.17592) # Values from Key Knowledge Documents for Endurance
water_co2_mix = manFL.FluidMix(fluid1=water, fluid2=co2)
brine_co2_mix = manFL.FluidMix(fluid1=brine, fluid2=co2)
```

```python
in_situ = {"Vp": 3.89, "Vs": 2.18, "Rho": 2.31} # Endurance reservoir values
water_co2_mix.saturation = 1
parameters = {
    "Phi": 0.25,
    "Km": 37.5,
    "fluid": water_co2_mix,
    "permeability": 1e-12,
    "patch_size": 0.0002,
    "bubble_radius": 0.0002,
    "Q_sls":15,
    "Log_omega_ref":0,
    "crack_density": 0.1,
    "aspect_ratio": 0.0001,
    "eps": 0.1,
    "tau":10
}

pars = {**in_situ, **parameters}
```

```python

models = {
    str(model.__name__).split(".")[-1]: {
        "name": model.Model.name,
        "type": model.Model.model_type.split(","),
    }
    for model in manRP.model_factory.ModelsBuild.models
}
calibrated = {model: manRP.models(identifier=model, **pars) for model in models}
calibrated["chapman"].eps = 0.005
for key, val in models.items():
    types = "\n".join(val["type"])
    print(f"{val['name']:-^40}\n{types}")
```

```python
s = np.linspace(0, 1, 100)
water_co2_mix.saturation = 1
water_co2_mix.patch_q = 1
#water_co2_mix.saturation = 0.88
#water_co2_mix.patch_q = co2.modulus/water.modulus
density = manRP.density.Density.from_insitu(Rho=in_situ["Rho"], fluid=water_co2_mix, Rho_d=2.65)

fig, ax = plt.subplots(1,2, figsize=(18,5))
data = np.empty((len(calibrated), len(s)))
att = data.copy()
for j, sat in enumerate(s):
    water_co2_mix.saturation = sat
    for i, model in enumerate(calibrated.items()):
        m, v = model
        data[i, j] = np.sqrt(np.real(v.Cij()[0,0] / density()))
        att[i, j]=np.abs(np.imag(v.Cij()[0,0]))/np.real(v.Cij()[0,0])
for i, d in enumerate(data):
    if i>4:
        dash = '--'
    else:
        dash = '-'
    ax[0].plot(s, d, label=calibrated[list(calibrated.keys())[i]].name, lw=2, linestyle=dash)
    ax[1].plot(
        s, att[i], label=calibrated[list(calibrated.keys())[i]].name, lw=2, linestyle=dash
    )

ax[0].set_xlim(0,1)
ax[1].set_xlim(0,1)
ax[0].set_xlabel("Saturation")
ax[0].set_ylabel("Vp (Km/s)")
ax[1].set_ylabel("Attenuation")
ax[1].set_xlabel("Saturation")

ax[1].legend(loc="upper right")
plt.show()
plt.close()
```
